```python
# dataset com o preço das plantações
path2 = kagglehub.dataset_download("srgiomanhes/stardew-valley-villagers-dataset")
csv_crops = f"{path2}/crops.csv"
```

```python
dfc = pd.read_csv(csv_crops)
dfc
```

```python
dfc.at[24, "Sell Price (Seed)"] = '200' # alterando dados do dataset original
dfc.at[3, "Name"] = "Coffee Bean"
dfc.at[3, "Sell Price (Seed)"] = '500'
```

```python
dfc.iloc[24]["Sell Price (Seed)"]
```

```python
precovenda = dfc['Price (Iridium)'].values
precobase = dfc['Sell Price (Seed)'].apply(int).values

precobase
```

```python
precobase = np.sort(precobase)
```

```python
plt.scatter(precovenda, precobase, color='r')
plt.xlabel('Preço da semente na compra (Pierre)')
plt.ylabel('Preço de venda (irídio)')
plt.title("Valor de venda x compra das sementes")
plt.grid(axis='both', ls='--', color='0.40')
plt.show()
```

```python
plt.xlim([-10, 650])
plt.ylim([-10, 110])
plt.scatter(precovenda, precobase, color='r')
plt.xlabel('Preço da semente na compra (Pierre)')
plt.ylabel('Preço de venda (irídio)')
plt.title("Valor de venda x compra das sementes")
plt.grid(axis='both', ls='--', color='0.40')
plt.show()
```

```python
tempocres = dfc['Growth Time (In Days)'].values
```

```python
plt.scatter(precovenda, precobase, c=tempocres, cmap='viridis')
plt.xlabel('Preço da semente na compra (Pierre)')
plt.ylabel('Preço de venda (irídio)')
plt.colorbar(label='Tempo de crescimento [dias]')
plt.title("Valor de venda x compra das sementes")
plt.grid(axis='both', ls='--', color='0.40')
plt.show()
```

```python
plt.scatter(precovenda, precobase, c=tempocres, cmap='viridis')
plt.xlabel('Preço da semente na compra (Pierre)')
plt.ylabel('Preço de venda (irídio)')
plt.xlim([-10, 610])
plt.ylim([-10,110])
plt.colorbar(label='Tempo de crescimento [dias]')
plt.title("Valor de venda x compra das sementes")
plt.grid(axis='both', ls='--', color='0.40')
plt.show()
```
